```python
import pandas as pd
df = pd.read_csv("netflix_titles.csv")
df.head()
```

```python
# Total rows aur columns
print("Shape of dataset:", df.shape)

# Columns ke names
print("\nColumns in dataset:")
print(df.columns)

# Dataset information (data types, null values, etc.)
print("\nDataset Info:")
df.info()

# Dataset ke basic statistics
print("\nDataset Description:")
df.describe(include='all')
```

```python
# Check missing values in each column
df.isnull().sum()
```

```python
# Drop column with too many missing values
df = df.drop(columns=['director'])

# Fill small missing values with 'Unknown'
df['cast'] = df['cast'].fillna('Unknown')
df['country'] = df['country'].fillna('Unknown')
df['date_added'] = df['date_added'].fillna('Unknown')
df['rating'] = df['rating'].fillna('Unknown')
df['duration'] = df['duration'].fillna('Unknown')

# Check again to confirm
df.isnull().sum()
```

```python
# Check duplicates
duplicates = df.duplicated().sum()
print("Total duplicate rows:", duplicates)

# Remove duplicates (if any)
df = df.drop_duplicates()

# Confirm again
print("After removing duplicates:", df.duplicated().sum())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Count of Movies vs TV Shows
plt.figure(figsize=(6,4))
sns.countplot(x='type', data=df, palette='Set2')
plt.title('Count of Movies vs TV Shows on Netflix')
plt.xlabel('Type')
plt.ylabel('Count')
plt.show()
```

```python
# Count top 10 countries
top_countries = df['country'].value_counts().head(10)

# Plot
plt.figure(figsize=(10,5))
sns.barplot(x=top_countries.values, y=top_countries.index, palette='coolwarm')

plt.title('Top 10 Countries with Most Content on Netflix')
plt.xlabel('Number of Titles')
plt.ylabel('Country')
plt.show()
```

```python
# Number of titles released each year
titles_per_year = df['release_year'].value_counts().sort_index()
# Plot
plt.figure(figsize=(12,6))
sns.lineplot(x=titles_per_year.index, y=titles_per_year.values, marker='o', color='purple')
plt.title('Number of Titles Released Each Year on Netflix')
plt.xlabel('Release Year')
plt.ylabel('Number of Titles')
plt.grid(True)
plt.show()
```

```python
# Count of each rating
rating_counts = df['rating'].value_counts().head(10)

# Plot
plt.figure(figsize=(10,5))
sns.barplot(x=rating_counts.values, y=rating_counts.index, palette='mako')
plt.title('Top 10 Content Ratings on Netflix')
plt.xlabel('Number of Titles')
plt.ylabel('Rating')
plt.show()
```

```python
# Step 7: Analyze the distribution of content types (Movies vs TV Shows)
type_count = df['type'].value_counts()  # Counts how many Movies and TV Shows are in the dataset
print(type_count)  # Prints the number of each type

# Visualize the distribution in a bar chart
type_count.plot(kind='bar', color=['skyblue', 'salmon'], title='Distribution of Content Types')  # Creates bar chart
plt.xlabel('Type')  # Label for X-axis
plt.ylabel('Count')  # Label for Y-axis
plt.show()  # Displays the chart
```

```python
# Step 10: Analyze most common genres/categories on Netflix
# We'll split the 'listed_in' column (which has multiple genres) and count the most frequent ones

from collections import Counter

# Split all genre strings into individual words and count their frequency
genre_list = df['listed_in'].dropna().apply(lambda x: [i.strip() for i in x.split(',')])  # Split by comma
all_genres = [genre for sublist in genre_list for genre in sublist]  # Flatten list
genre_count = Counter(all_genres)  # Count each genre
print(genre_count)
# Convert to DataFrame for easier plotting
genre_df = pd.DataFrame(genre_count.most_common(10), columns=['Genre', 'Count'])

# Plot the top 10 genres
plt.figure(figsize=(10,6))
sns.barplot(x='Count', y='Genre', data=genre_df, palette='coolwarm')

plt.title('Top 10 Most Common Genres on Netflix')
plt.xlabel('Number of Titles')
plt.ylabel('Genre')
   # Highest at top
plt.show()
```

```python
# Step 11: Trend of Netflix content release over the years

# Count number of movies and TV shows released each year
content_trend = df.groupby(['release_year', 'type']).size().unstack(fill_value=0)

# Plot trend
plt.figure(figsize=(12,6))
content_trend.plot(kind='line', marker='o')
plt.title('Netflix Content Released per Year')
plt.xlabel('Release Year')
plt.ylabel('Number of Titles')
plt.legend(title='Type')
plt.grid(True)
plt.show()
```
